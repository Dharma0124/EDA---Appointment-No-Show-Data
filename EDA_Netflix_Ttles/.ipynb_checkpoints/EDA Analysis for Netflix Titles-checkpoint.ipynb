# EDA Analysis for Netflix Titles

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno  # For missing data visualization
#!pip install missingno
# Configure visualizations
%matplotlib inline
sns.set(style="whitegrid")
```

## Reading the Dataset

```python
# Load the dataset
data = pd.read_csv('C:\\Users\\dharm\\Downloads\\netflix_titles_2021.csv')
# Quick view of the dataset
data.head()
```

## Initial Inspection

```python
#check dimensions
print("Dataset Dimensions:", data.shape)
#Data summary:
data.info()
```

## Checking for Duplicates

```python
# Checking for duplicates
print("Number of duplicate rows:", data.duplicated().sum())
```

```python
#Remove duplicates:
data = data.drop_duplicates()
```

## Missing Values Calculation

```python
# Check for missing values
print(data.isnull().sum())
# Visualize missing data
msno.matrix(data)
plt.show()
```

## Data Reduction

```python
# Dropping columns (example)
data = data.drop(columns=['show_id'])  # if show_id is not needed
```

## Feature Engineering

```python
# Strip whitespace and convert `date_added` to datetime
data['date_added'] = pd.to_datetime(data['date_added'].str.strip(), errors='coerce')

# Extracting year and month from `date_added`
data['year_added'] = data['date_added'].dt.year
data['month_added'] = data['date_added'].dt.month
```

## Creating Features

```python
# Extract numerical duration value
data['duration_num'] = data['duration'].str.extract(r'(\d+)').astype(float)
```

## Data Cleaning/Wrangling

```python
# Clean 'type' column values
data['type'] = data['type'].str.strip().str.lower()
```

## EDA - Exploratory Data Analysis (Univariate)

```python
# Plotting distributions of categorical features
sns.countplot(data['type'])
plt.title('Content Types on Netflix')
plt.show()

# Histogram of content added by year
sns.histplot(data['year_added'].dropna(), bins=15, kde=True)
plt.title('Distribution of Content Added by Year')
plt.show()
```

## Statistics Summary

```python
# Summary statistics for numerical columns
data.describe()
# Data Transformation
# Example: Converting year added to categorical
data['year_added'] = data['year_added'].astype('category')
```

## EDA - Bivariate Analysis

```python
# Bivariate analysis between type and rating
sns.countplot(x='rating', hue='type', data=data)
plt.title('Distribution of Ratings by Content Type')
plt.xticks(rotation=45)
plt.show()

# Relationship between duration and year added
sns.scatterplot(x='year_added', y='duration_num', hue='type', data=data)
plt.title('Content Duration by Year Added')
plt.show()
```

## EDA - Multivariate Analysis

```python
# Correlation matrix for numerical values
correlation_matrix = data.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Netflix Data")
plt.show()
```

## Impute Missing Values

```python
# Replace missing values in the 'rating' column
data['rating'] = data['rating'].fillna('Not Rated')

# Replace missing values in the 'year_added' column
data['year_added'] = data['year_added'].fillna(data['year_added'].mode()[0])
```
